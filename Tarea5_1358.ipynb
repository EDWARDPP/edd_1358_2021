```
class NodoDoble:
  def __init__(self, value, previo = None, siguiente = None):
    self.data = value
    self.prev = previo
    self.next = siguiente

class DoubleLinkedList:
  def __init__(self):
    self.__head = None
    self.__tail = None
    self.__size = 0 #Para obtener el tama√±o de la lista 

  def get_size(self):
    return self.__size

  def is_empty(self):
    return self.__size == 0
  
  def append(self, value):
    if self.is_empty():
      nuevo = NodoDoble(value)
      self.__head = nuevo
      self.__tail = nuevo
    else:
      nuevo = NodoDoble(value, self.__tail, None)
      self.__tail.next = nuevo
      self.__tail = nuevo

    self.__size += 1

  def find_from_head(self, value ):
    curr_node = self.__head
    if self.__head == None: #Validando si la lista esta vacia 
      return self.is_empty()
    
    elif self.__head.data == value: #Validando si la cabeza es el valor solicitado
      return 0
    else:
     pos = 0
     while curr_node.data != value:
      curr_node = curr_node.next
      pos +=1
    if curr_node.data == value:
      return pos
    

  def find_from_tail(self, value ):
     curr_node = self.__tail
     if self.__tail == None:   #Validando si la lista esta vacia 
      return self.is_empty()
    
     elif self.__tail.data == value: #Validando si la cola es el valor solicitado
      return 0
      return 0
     else:
      pos = 0
      while curr_node.data != value:
       curr_node = curr_node.prev
       pos +=1
     if curr_node.data == value:
      return pos
    


    
  def remove_from_head(self, value):
    if self.is_empty():
      pass

    if self.__head.data == value: #Validando si la cabeza es el dato que se quiere eliminar
      self.__head = self.__head.next
      self.__head.prev = None
      return

    if self.__tail.data == value: #Validando si la cola es el dato que se quiere eliminar
      self.__tail = self.__tail.prev
      self.__tail.next = None
    else:

     curr_node = self.__head
     while curr_node.data != value and curr_node != None:
        curr_node = curr_node.next
     if curr_node.data == value: 
         curr_node.prev.next = curr_node.next
         curr_node.next.prev = curr_node.prev
     
    self.__size-= 1

  def remove_from_tail(self, value):
    if self.is_empty():
      pass
    if self.__tail.data == value: #Validando si la cola es el dato que se quiere eliminar
      self.__tail = self.__tail.prev
      self.__tail.next = None
      return
    
    if self.__head.data == value: #Validando si la cabeza es el dato que se quiere eliminar
      self.__head = self.__head.next
      self.__head.prev = None
    else:

     curr_node = self.__tail
     while curr_node.data != value and curr_node != None:
        curr_node = curr_node.prev
     if curr_node.data == value: 
         curr_node.next.prev = curr_node.prev
         curr_node.prev.next = curr_node.next
     
    self.__size-= 1
      

  def transversal(self):
    curr_node = self.__head
    while curr_node != None:
        print("<----", curr_node.data, "---->", end="")
        curr_node = curr_node.next
    print("")

  def reverse_transversal(self):
    curr_node = self.__tail
    while curr_node != None:
        print("<----", curr_node.data, "---->", end="")
        curr_node = curr_node.prev
    print("")
    
```
