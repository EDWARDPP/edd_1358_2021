```
#Clase cola.
class Queue:
  def __init__(self):
    self.__data = list()

  def is_empty(self):
    return len(self.__data) == 0
  
  def length(self):
    return len(self.__data)

  def enqueue(self, item):
    self.__data.append(item)

  def dequeue(self):
    if not self.is_empty():
      return self.__data.pop(0)
    else: 
      return None

  def to_string(self):
    cadena = ""
    for elem in self.__data:
      cadena = cadena+"|"+ elem 
    cadena = cadena + "|" 
    return cadena
```

```
#Clase cola de prioridad no acotada
class NoAcotadaPriorityQueue:
  def __init__(self):
    self.__data= list()
    self.__size =0

  def is_empty(self):
    return self.__size == 0 

  def length(self):
    return self.__size

  def enqueue(self, prioridad, elem):
    if self.is_empty() or prioridad > self.__size:
      self.__data.append([elem,prioridad])
      self.__size += 1
    else:
      a = self.__size -1 
      while self.__data[a][1] != prioridad and a >= 0:
        a -= 1
      if a != -1: 
        self.__data.insert((a+1),[elem,prioridad]) 
        self.__size += 1
      else:
        if prioridad > self.__data[self.__size -1][1]:
          self.__data.append([elem,prioridad])
          self.__size += 1
        else:
          b = self.__size -1 
          while prioridad < self.__data[b][1] and b >= 0:
            b-= 1
          self.__data.insert((a),[elem,prioridad]) 
          self.__size += 1
  
  
  def dequeue(self):
     self.__size -=1
     self.__data.pop(0)
   

  def nato_string(self):
     for i in range(len(self.__data)) :
       print(f"| {self.__data[i][0]}")
     print("|")
```

```
#Clase de prioridad acotada
class BoundedPriorityQueue:
  def __init__(self, niveles):
    self.__data = [Queue() for x in range(niveles)]
    self.__size = 0
  
  def is_empty(self):
    return self.__size == 0

  def length(self):
    return self.__size

  def enqueue( self, prioridad, elem):
    if prioridad < len(self.__data) and prioridad >= 0:  #Valida que la prioridad no sea mayor al numero de elemntos de la lista.
      self.__data[prioridad].enqueue(elem) 
      self.__size +=1
    else:
      print ("Prioridad fuera de rango")

  def dequeue(self):
    if not self.is_empty():
      nivel = 0
      while self.__data[nivel].is_empty():
        nivel+=1 
      self.__size-=1
      return self.__data[nivel].dequeue()
      
        

  def to_string(self):
    for nivel in range(len(self.__data)):
      print(f"Nivel {nivel} ---> {self.__data[nivel].to_string() }")

```
